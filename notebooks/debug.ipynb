```python
import pickle

with open("docs-all-chunks", "rb") as file:
    docs = pickle.load(file)
```

```python
docs[0].chunks.keys()
```

```python
for doc in docs:
    print(doc.filename)
    for strat in doc.chunks.keys():
        print(strat, ":", len(doc.chunks[strat]))
        
```

```python
for key in docs[0].chunks.keys():
    print("#", key.upper(), "\n")
    print(len(docs[0].chunks[key]), "chunk(s)\n")
    for chunk in docs[0].chunks[key]:
        print(chunk, "\n")
    # print(docs[0].chunks[key])
    print("---")
```

```python
# output all chunks
for doc in docs:
    output = f"# {doc.filename}\n\n"
    for key in doc.chunks.keys():
        output += f"## {key.upper()}\n\n"
        output += f"#### {len(doc.chunks[key])} chunk(s)\n\n"
        for chunk in doc.chunks[key]:
            output += f"{chunk}\n\n"
            # print(chunk, "\n")
        # print(docs[0].chunks[key])
        # print("---")   
    output += "---"

    # print(output)

    with open(f"chunk_texts/{doc.filename}.md", "w") as file:
        file.write(output)
    
# for key in docs[0].chunks.keys():
#     output += f"## {key.upper()}\n"
#     output += f"{len(doc.chunks[key])} chunk(s)\n"
#     for chunk in doc.chunks[key]:
#         output += f"{chunk}\n\n"
#         # print(chunk, "\n")
#     # print(docs[0].chunks[key])
#     print("---")   
#     output += "---"
```
